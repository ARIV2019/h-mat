# 3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.
Напишите код, который будет рисовать график окружности в полярных координатах.
Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.


```python
# 1. Перевод из полярных в декартовы + график окружности 
phi = np.linspace(0, 2*np.pi, 1000)
rho = 10
x = rho * np.cos(phi)
y = rho * np.sin(phi)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# 2. график окружности в полярных координатах
t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
rho = np.sqrt(x**2 + y**2)
phi = np.arctan2(y, x)*180/np.pi
plt.polar(phi, rho)
```

```python
def xy2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)*180/np.pi
    return(rho, phi)

t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
rho, phi = xy2pol(x, y)
plt.polar(phi, rho)
```

```python
# 3. график прямой 
t = np.linspace(0, np.pi, 1000)
x = np.sin(t)
y = np.sin(t)
rho = np.sqrt(x**2 + y**2)
phi = np.arctan2(y, x)*180/np.pi
plt.polar(phi, rho)
```

# 4. Задание (в программе)
Решите систему уравнений:

#  Пример 1: 
## $y = x^2 - 1$
## $exp(x) + x (1 - y) = 1$

```python
from scipy.optimize import fsolve
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) - 1)/x + 1)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,7) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 =  fsolve(equations, (-2, 1))
print (x1, y1)

x2, y2 =  fsolve(equations, (2, 2))
print (x2, y2)
```
