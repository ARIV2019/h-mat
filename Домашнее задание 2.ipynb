# # 2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)



```python
x = np.linspace(-5, 5, 21)
y = 3*x + 1
y2 = (-1/3)*x + 1
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```

#  Прямые на графике не кажутся перпендикулярными, потому чно они перпендикулярны в пространстве, а не на плоскости. Две прямые в пространстве перпендикулярны,если угол между ними 90 гр.  На плоскости перпендикулярными могут оказаться только пересекающиеся прямые. А перпендикулярность в пространстве двух прямых может быть даже в случае если они не пересекаются

# # 5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.

```python
fig = plt.figure(num=1, clear=True)
ax = fig.add_subplot(1, 1, 1, projection='3d')


(x, y) = np.meshgrid(np.arange(-2, 2.1, 1), np.arange(-2, 2.1, 1))
z = x + y

ax.plot_surface(x, y, z, color='blue')
ax.plot_surface(x, y, z, color='yellow')
ax.set(xlabel='x', ylabel='y', zlabel='z', title='z = x + y')

fig.tight_layout()
```
